Récupérer une page Wikipédia (celle sur la suisse) avec la bibliothèque Requests et l'enregistrer dans un fichier texte

```python
# 1. j'importe requests
import requests

# 2. je fais une requête sur la page wikipedia de la suisse
suisse = requests.get("https://fr.wikipedia.org/wiki/Suisse")

# 3. et je stocke le contenu text
contenu_texte = suisse.text
```

```python
# vérifions si on a bien récupéré ce que l'on veut
contenu_texte
```

```python
# 4. j'utilise open en écriture -> "w" pour créer un fichier wiki_suisse
with open("wiki_suisse.txt", "w") as fichier:
    # 5. je demande d'écrire le contenu_texte dans le fichier créé
    fichier.write(contenu_texte)

# un fichier a bien été créé dans mon répertoire.
```

```python
# si jamais je veux vérifier :

# j'ouvre le fichier en question en mode lecture -> "r" puis j'affiche son contenu
with open("wiki_suisse.txt","r") as fichier:
    contenu_fichier = fichier.read()
    
contenu_fichier
```
